```python
# 전체 평균 수면시간 보다 이상이면 1, 미만이면 0
def update_above_auas(row):
    if row["Avg sleep time"] >= row["AllUserAvgSleep"]:
        return 1
    else:
        return 0


temp_new_train_events_allAVG["AboveAUAS"] = temp_new_train_events_allAVG.apply(
    update_above_auas, axis=1
)
temp_new_train_events_allAVG
```

```python
# 이벤트 수면 시작 : 1 ,  수면 종료 : 0
def event_invert(row):
    if row["event"] == "onset":
        return 1
    else:
        return 0


temp_new_train_events_allAVG["event"] = temp_new_train_events_allAVG.apply(
    event_invert, axis=1
)
temp_new_train_events_allAVG
```
