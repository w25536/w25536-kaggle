```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns


import os
from pathlib import Path

BASE_DIR = Path.cwd().parent

df = pd.read_csv(os.path.join(BASE_DIR, 'csv', 'admission_predict.csv'), index_col= 'Serial No.')
```

```python
df
```

```python
X = df.drop('Chance of Admit ', axis = 1)
y = df['Chance of Admit ']
```

```python
print(X.shape)
print(y.shape)
print(type(X))
print(type(y))
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.fit_transform(X)
```

```python
X =pd.DataFrame(X)
#y = pd.DataFrame(y)
```

```python
y = y.reset_index(drop=True)

y = pd.DataFrame(y)
```

```python
df = pd.concat([X, y], axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
from pycaret.regression import setup, compare_models

setup(data=X_train, target=y_train, session_id=42)
```

```python
compare_models()
```

```python
from sklearn.linear_model import ridge_regression
ridge_model = ridge_regression(alpha=1.0, random_state=42)

ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
```

```python
ridge_model.score(X_test, y_test)
```

```python
plt.figrue(figsize=(14,10))
plt.plot(y_pred, y_test, 'o')
```
